```python
import pandas as pd 
df=pd.read_csv('football_matches.csv')
```

```python
df
```

```python
# Which team scored the most total goals 
df['total_goals'] = df['goal_home_ft'] + df['goal_away_ft']
home_goals = df.groupby('home_team')['goal_home_ft'].sum()
away_goals = df.groupby('away_team')['goal_away_ft'].sum()
total_team_goals = (home_goals + away_goals).sort_values(ascending=False)
print(total_team_goals.head(10))
```

```python
#Which match had the most goals?
df['total_goals'] = df['goal_home_ft'] + df['goal_away_ft']
max_goals_match = df.loc[df['total_goals'].idxmax()]
print(max_goals_match[['date', 'home_team', 'away_team', 'goal_home_ft', 'goal_away_ft', 'total_goals']])
```

```python
#Average goals per match
avg_goals = df['total_goals'].mean()
print(f"Average goals per match: {avg_goals:.2f}")
```

```python
#Most attacking teams (based on total shots)
df['total_shots'] = df['home_shots'] + df['away_shots']
team_shots = df.groupby('home_team')['home_shots'].sum() + df.groupby('away_team')['away_shots'].sum()
print(team_shots.sort_values(ascending=False).head(10))
```

```python
# Most aggressive teams (based on fouls + yellow + red cards)
df['home_aggression'] = df['home_fouls_conceded'] + df['home_yellow_cards'] + df['home_red_cards']
df['away_aggression'] = df['away_fouls_conceded'] + df['away_yellow_cards'] + df['away_red_cards']
aggression = df.groupby('home_team')['home_aggression'].sum() + df.groupby('away_team')['away_aggression'].sum()
print(aggression.sort_values(ascending=False).head(10))
```

```python
#Top 5 matches where underdog (weaker team) won (based on shots)
df['underdog_win'] = ((df['home_shots'] < df['away_shots']) & (df['result'] == 1)) | \
                     ((df['home_shots'] > df['away_shots']) & (df['result'] == -1))
print(df[df['underdog_win'] == True][['home_team', 'away_team', 'home_shots', 'away_shots', 'result']].head(5))
```
